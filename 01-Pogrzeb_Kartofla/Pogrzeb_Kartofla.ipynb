```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
import matplotlib
%matplotlib inline
```

```python
for i in [1,2,3,4]:
    #Pobieranie danych
    SymCze = pd.read_csv(f"./Dane/Ziemniak{i}SymCze.bod",sep=r'\s*', skiprows=15, header=None) 
    SymCze.columns = ["Frequency","Gain(dB)","Gain","Phase(deg)"]
    
    #wykresy dla symulacji
    plt.figure(figsize=(10,8))
    plt.plot(SymCze['Frequency'],SymCze['Gain(dB)'],'c',label='punkty pomiarowe')
    plt.xscale('log')
    plt.grid()
    plt.title(f"Symulacja układu {i} \n Zależność $|k_u|$ od częstotliwości")
    plt.xlabel("f[Hz]")
    if (i != 4):
        f = interp1d(SymCze['Gain(dB)'],SymCze['Frequency'])
        f_g = f(max(SymCze['Gain(dB)'])-3)
        plt.axvline(f_g,color='C2',linestyle='dashed',label='częstotliwość graniczna')
        plt.legend()
        print(f'Dla ukladu {i} częstotliwosc graniczna to {f_g}')
    plt.ylabel("$|k_u|[dB]$")
    plt.savefig(f"Sym{i}CzeKu.png",dpi=300)
    plt.figure(figsize=(10,8))
    plt.plot(SymCze['Frequency'],SymCze['Phase(deg)'],'c')
    plt.xscale('log')
    plt.grid()
    plt.title(f"Symulacja układu {i} \n Zależność przesunięcia fazowego od częstotliwości")
    plt.xlabel("f[Hz]")
    plt.ylabel("$\Delta \phi [deg]$") 
    plt.savefig(f"Sym{i}CzeFaza.png",dpi=300)
```

```python
zbocze = [20,-20,-20]
przesuniecie = [-60,60,40]
f_g_teoria = [1040.23,1052,1156]

for i in [1,2,3,4]:
    #Pobieranie danych
    ElvisCze = pd.read_csv(f"./Dane/Pogrzeb{i}Cze.txt",sep=r'\s*', skiprows=3, header=None, decimal=',')
    ElvisCze.columns = ["Freq(Hz)","Gain(dB)","Phase(deg)"]
    
    #Wykresy dla Elvisa
    plt.figure(figsize=(10,8))
    plt.plot(ElvisCze['Freq(Hz)'],ElvisCze['Gain(dB)'],'.c',label='punkty pomiarowe')
    if(i != 4):
        f = interp1d(ElvisCze['Gain(dB)'],ElvisCze['Freq(Hz)'])
        x = [ElvisCze['Freq(Hz)'][0],f(max(ElvisCze['Gain(dB)'])-3),ElvisCze['Freq(Hz)'].iloc[-1]]
        y = [ElvisCze['Gain(dB)'][0],max(ElvisCze['Gain(dB)']),ElvisCze['Gain(dB)'].iloc[-1]]
        plt.plot(x,y,'C1',label='liniowa aproksymacja')
        plt.legend()
        f_g = f(max(ElvisCze['Gain(dB)'])-3)
        plt.axvline(f_g,color='C2',linestyle='dashed',label='częstotliwość graniczna')
        print(f'Dla ukladu {i} częstotliwosc graniczna to {f_g}')
    plt.xscale('log')
    plt.grid()
    plt.title(f"Układ doświadczalny {i} \n Zależność $|k_u|$ od częstotliwości")
    plt.xlabel("f[Hz]")
    plt.ylabel("$|k_u|[dB]$")
    plt.savefig(f"Elvis{i}CzeKu.png",dpi=300)
    plt.figure(figsize=(10,8))
    plt.plot(ElvisCze['Freq(Hz)'],ElvisCze['Phase(deg)'],'.c')
    plt.xscale('log')
    plt.title(f"Układ doświadczalny {i} \n Zależność przesunięcia fazowego od częstotliwości")
    plt.xlabel("f[Hz]")
    plt.grid()
    plt.ylabel("$\Delta \phi [deg]$")
    plt.savefig(f"Elvis{i}CzeFaza.png",dpi=300)
```

```python
for i in [1,2,3,4]:
    for j in [1,2]:
        SymT = pd.read_csv(f"./Dane/Ziemniak{i}SymT{j}.bod",sep=r'\s*', skiprows=19, header=None)
        SymT.columns = ["Time(s)","Channel_A(V)","Channel_B(V)"]
        if (j==1):
            f = 100
        if (j==2):
            f = 10000
        fig, (ax1,ax2) = plt.subplots(2, sharex=True, figsize=(10,8))
        ax1.plot(SymT['Time(s)'],SymT['Channel_A(V)'],'c',label='$U_1$')
        ax2.plot(SymT['Time(s)'],SymT['Channel_B(V)'],'C1',label='$U_2$')
        ax1.grid()
        ax2.grid()
        ax1.set_title(f"Symulacja układu {i} \n Częstotliwość: {f}Hz \n Wykres sygnałów wejściowego i wyjściowego od czasu")
        ax2.set_xlabel("t[s]")
        ax1.set_ylabel("U1[V]")
        ax2.set_ylabel("U2[V]")
        fig.savefig(f"Sym{i}T{j}.png",dpi=300)
```

```python
for i in [1,2,3,4]:
    for j in [1,2]:
        ElvisT = pd.read_csv(f"./Dane/Pogrzeb{i}T{j}.txt",sep=r'\s*', skiprows=5, header=None, parse_dates=[[0,1],[3,4]])
        ElvisT.columns = ["time0","time1","y0",'y1']
        ElvisT['y0'] = pd.to_numeric(ElvisT['y0'].str.replace(',','.'))
        ElvisT['y1'] = pd.to_numeric(ElvisT['y1'].str.replace(',','.'))

        if (j==1):
            f = 100
        if (j==2):
            f = 10000
        fig, (ax1,ax2) = plt.subplots(2, sharex=True, figsize=(10,8))
        ax1.plot((ElvisT['time0']-ElvisT['time0'].iloc[0])/1e3,ElvisT['y0'],'.-c',label='$U_1$')
        ax2.plot((ElvisT['time1']-ElvisT['time1'].iloc[0])/1e3,ElvisT['y1'],'.-C1',label='$U_2$')
        ax1.grid()
        ax2.grid()
        ax1.set_title(f"Układ doświadczalny {i} \n Częstotliwość: {f}Hz \n Wykres sygnałów wejściowego i wyjściowego od czasu")
        ax2.set_xlabel(r"t[$\mu $s]")
        ax1.set_ylabel("U1[V]")
        ax2.set_ylabel("U2[V]")
        fig.savefig(f"Elvis{i}T{j}.png",dpi=300)
```
