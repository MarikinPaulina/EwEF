```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sympy as sp
from sympy import inverse_laplace_transform
from scipy.interpolate import interp1d
%matplotlib inline
sp.init_printing('latex')
from IPython.display import display
```

```python
L = 18.16e-3
R = 50
R_g = 50
r = 0.46
C = 2.2e-6
```

```python
s, t = sp.symbols('s, t')
k_s = 1/(1+s*C*(R+R_g+r)+L*C*s**2)
k_t = inverse_laplace_transform(k_s,s,t)
print(k_t)
t = np.linspace(0,5e-3,1000)
def k_t(t):
    return 6004.04*np.exp(-2765.97*t)*np.sin(4168.87*t)
plt.plot(t*1e3,k_t(t),'k')
plt.xlabel("t[ms]")
plt.ylabel("k(t)")
plt.grid()
plt.title('Odpowiedź na sygnał impulsowy')
plt.savefig("kt.png",dpi=400)
```

```python
omega = sp.symbols('omega')
k = sp.lambdify(omega, sp.Abs(k_s.subs({s: sp.I * omega})), 'numpy')

om = np.logspace(1,6,1000)
# def h_t(t):
#     return 6004.04*np.exp(-2765.97*t)*np.sin(4168.87*t)
plt.loglog(om/(2*np.pi),k(om),'k')
f = interp1d(20*np.log10(k(om)),om/(2*np.pi))
f_g = f(-3)
plt.xlabel("f[Hz]")
plt.ylabel(r"$k(f)$")
plt.grid()
plt.title("Charakterystyka częstotliwościowa")
plt.savefig("charakterystyka_czestotliwosciowa.png", dpi=400)
f_g
```

```python
s = sp.symbols('s')
t = sp.symbols('t', positive=True)
h_s = 1/((1+s*C*(R+R_g+r)+L*C*s**2)*s)
# h_t = inverse_laplace_transform(h_s,s,t)
h_t_sympy = 1 - (0.5 - 0.331741j)*sp.exp((-2765.97 - 4168.87j)* t) * ((0.388688 + 0.92137j) + (1)* sp.exp((8337.73j)*t))
display(h_t)
h_t = sp.lambdify(t, h_t_sympy, 'numpy')
tRIGGERED = np.linspace(0,5e-3,1000)
# def h_t(t):
#     return 6004.04*np.exp(-2765.97*t)*np.sin(4168.87*t)
plt.plot(tRIGGERED*1e3,h_t(tRIGGERED).real,'k')
plt.xlabel("t[ms]")
plt.ylabel("h(t)")
plt.grid()

plt.title('Odpowiedź na skok jednostkowy')
plt.savefig("ht.png",dpi=400)
```

```python
h_re, h_im = h_t_sympy.as_real_imag()
nowa_lepsza_h_t = sp.lambdify(t, h_re, 'numpy')
def h_t_takie_sobie(t):
    return 1 - np.exp(-2765.97*t)*(0.663482*np.sin(4168.87*t)+np.cos(4168.87*t))
h_re.simplify().expand()
```

```python
plt.plot(tRIGGERED*1e3,h_t(tRIGGERED),'k')
plt.plot(tRIGGERED*1e3,nowa_lepsza_h_t(tRIGGERED))
plt.plot(tRIGGERED*1e3,h_t_takie_sobie(tRIGGERED))
```
